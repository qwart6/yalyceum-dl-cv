# Градиентный спуск по неизвестной функции

```python
import numpy as np


def gradient_descent(func, start_point, gamma, epsilon, steps):
    delta_x = 1e-9
    current_x = float(start_point)
    history = [round(current_x, 3)]

    if steps > 0:
        for _ in range(steps):
            y_old = func(current_x)
            grad = (func(current_x + delta_x) - y_old) / delta_x
            current_x -= gamma * grad
            history.append(round(current_x, 3))
    else:
        while True:
            y_old = func(current_x)
            grad = (func(current_x + delta_x) - y_old) / delta_x
            new_x = current_x - gamma * grad
            y_new = func(new_x)
            current_x = new_x
            history.append(round(current_x, 3))
            if abs(y_new - y_old) < epsilon:
                break

    return np.array(history).reshape(-1, 1)
```

```python
def f2(x):
    return x**2 + 5*x + 6

result2 = gradient_descent(f2, start_point=5.0, gamma=0.1, epsilon=1e-6, steps=30)
print("f(x) = x^2 + 5x + 6")
print(f"Траектория: ")
print(f"{result2.flatten()}")
print(f"Конечная точка: {result2[-1][0]}")
print()
```

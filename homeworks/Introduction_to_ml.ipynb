## Метрики регрессии

```python
import numpy as np


def metrics_regretion(target, pred):
    mse = np.sum((np.array(target) - np.array(pred)) ** 2) / len(target)
    mae = np.sum(np.abs(np.array(target) - np.array(pred))) / len(target)
    rmse = np.sqrt(mse)

    print(f'MSE: {mse:.2f}\nMAE: {mae:.2f}\nRMSE: {rmse:.2f}')


target = list(map(float, input().split())) # 1 5 2 9 2 3
pred = list(map(float, input().split()))   # 1.5 5.2 2.3 8.7 2.8 3.3

metrics_regretion(target, pred)
```

## Коэффициент детерминации

```python
import numpy as np


def r2_score(target, pred):
    target = np.array(target)
    pred = np.array(pred)

    mean_target = np.mean(target)

    r2 = 1 - (np.sum((target - pred) ** 2) / (np.sum((mean_target - target) ** 2)))

    print(f'R2: {r2:.2f}')


target = list(map(float, input().split())) # 1 5 2 9 2 3
pred = list(map(float, input().split()))   # 1.5 5.2 2.3 8.7 2.8 3.3

r2_score(target, pred)
```

## KNN на практике

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

data_frame = pd.read_csv('/Users/artemvorobchenko/Downloads/penguins (2).csv')

data_frame.dropna(inplace=True)

labels = data_frame['species']
features = data_frame.iloc[:, 2:4]

train_features, test_features, train_labels, test_labels = train_test_split(
    features, labels, test_size=0.2, random_state=123)

accuracy_list = list()

for neighbors in range(1, 10 + 1):
    for weight in ['uniform', 'distance']:
        knn_model = KNeighborsClassifier(n_neighbors=neighbors, weights=weight)
        knn_model.fit(train_features, train_labels)

        pred = knn_model.predict(test_features)
        accuracy = accuracy_score(test_labels, pred)
        accuracy_list.append(accuracy)

print('Best accuracy: {:.6f}'.format(max(accuracy_list)))
print('Worst accuracy: {:.6f}'.format(min(accuracy_list)))
```
